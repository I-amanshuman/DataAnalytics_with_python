```python
#Now let us look into multivariate plots where we will plot multiple factors to determine whether one has any impact on the other
#or not

#Plot primary sales in the primary y-axes and radio TRPs in the secondary y-axes and determine if they are correlated

fig, ax1 = plt.subplots() #Initiating a subplot to create 2 objects, fig and ax1, figure element and axes element respectively
ax1.plot(data3['NewVolSales'])
ax1.set_xlabel('time in weeks')
ax1.set_ylabel('Volume sales')

#Create a secondary y axes

data3.columns

ax2 = ax1.twinx()
ax2.plot(data3['Radio '], color = 'red')
ax2.set_ylabel('Radio')

fig.tight_layout()
plt.show()
```

```python
#Scatter plot: Bivariate plot
#Relationship between age and income for data3

plt.scatter(x=data['Income'], y=data['Age'], color = 'lightblue', alpha = 0.5)
plt.xlabel('Income')
plt.ylabel('Age')

plt.show()

#positive relationship between age and income
```

## Using seaborn for faceted plots/ plotting by groups

```python
#seaborn is built on matplotlib, so we need to map the same

import seaborn as sns
sns.FacetGrid(col='Department', data=data, size = 5).map(plt.scatter, 'Age', 'Income')

```

```python
#Look at the distribution of Income by Gender for different Marital Status, ie plot the faceted chart on 2 diff rows
```

```python
sns.FacetGrid(col = 'MaritalStatus', row = 'Gender', data = data, size = 4).map(plt.hist, 'Income', bins = 20, color = 'm')
```

## Using subplots for multiple plots

```python
plt.subplot(1,2,1) #Create a figure with 1x2 subplot and make the 1st subplot active, ie the left one
plt.scatter(data['Age'], data['Income'], color = 'b')
plt.xlabel('Scatter')

plt.subplot(1,2,2)
plt.boxplot(data.Income)
plt.xlabel('Boxplot')

plt.tight_layout()
plt.show()
```
